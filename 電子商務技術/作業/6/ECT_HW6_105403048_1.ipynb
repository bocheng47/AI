```python
import pandas as pd
#讀取CSV檔案
data = pd.read_csv('BreastCancer.csv')
```

# (a). 將 radius_mean 及 area_mean 切為 feature，diagnosis 切為 target

```python
Target = data['diagnosis']

#將屬性合併
#變成list
feature=list(zip(data['radius_mean'],data['area_mean']))

import numpy as np
#轉成array
features=np.asarray(feature)

from sklearn import preprocessing
#轉換屬性型態
#將屬性轉為數字label
le = preprocessing.LabelEncoder()
#將 diagnosis 轉為數字 diagnosis 
Target=le.fit_transform(Target)
```

# (b). 切分資料及與訓練集，設 test_size=0.34，random_state=5

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature, Target, test_size=0.34, random_state=5)
```

# (c). 用 KNeighborsClassifier 進行 KNN 分類，n_neighbors 設為 6

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=6)
neigh.fit(X_train, y_train) 
y_result = neigh.predict(X_test) 
```

# (d). 用 metrics 算出此模型對於測試集的準確度

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, y_result)
print("準確率為 :　%f" % score)
```

# (e). 運用 matplotlib 中的 scatter 圖，x 軸設為 radius_mean，y 軸設為area_mean，c 設為 label，印出測試分類圖形。

```python
import matplotlib.pyplot as plt
X = data[['radius_mean','area_mean']].values
label = np.arctan2(X[:, 1], X[:, 0])
plt.scatter(X[:, 0], X[:, 1], s=25, c=label, alpha=1)
```

# (f). 用 cluster.Means 設 n_clusters=2

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
```

# (g). 用 fit_predict 對切分的 feature 進行預測

```python
# 使用 K-Means 演算法
clusters = kmeans.fit_predict(X)
```

# (h). 運用 matplotlib 中的 scatter 圖，x 軸設為 radius_mean，y 軸設為area_mean，c 設為分群結果，印出分類圖形。

```python
plt.scatter(X[:, 0], X[:, 1], s=25, c=clusters, alpha=1)
```
